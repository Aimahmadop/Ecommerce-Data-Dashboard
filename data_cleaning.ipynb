```python
df = pd.read_csv('online_retail.csv')
df.head()
```

```python
df[['Quantity','UnitPrice']].describe()
```

```python
df.isnull().sum()
```

```python
#We are going to fill the missing with "99999" 
df['CustomerID']=df['CustomerID'].fillna(99999)
print("Remaining missing values is CUstomerID=",df['CustomerID'].isnull().sum())
```

```python
df['Description'] = df['Description'].fillna("Unknown Item")
print("The remaining missing values in Description is =", df['Description'].isnull().sum())
```

```python
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
df[['Quantity','UnitPrice']].describe()
```
